```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
movie = pd.read_csv("movies.csv")
rating = pd.read_csv("ratings.csv")
```

```python
movie.head()
```

```python
rating.head()
```

```python
movie_final = pd.merge(movie,rating,on="movieId").drop(["genres","timestamp","movieId"],
axis=1)
```

```python
movie_final["userId"] = "User"+movie_final["userId"].astype("str")
```

```python
movie_final
```

```python
from surprise import Dataset, SVD, accuracy, Reader
from surprise.model_selection import train_test_split
```

```python
movie_final1 = movie_final
```

```python
reader = Reader(rating_scale=(1,5))
```

```python
data = Dataset.load_from_df(movie_final,reader)
```

```python
trainset,testset = train_test_split(data,test_size=0.3,random_state=1)
```

```python
svd = SVD(n_factors=100)
```

```python
svd.fit(trainset)
```

```python
predictions = svd.test(testset)
accuracy.rmse(predictions)
```

```python
movie_user_matrix = movie_final1.pivot_table(index="userId",columns="title",values="rating")
```

```python
movie_user_matrix.head()
```

```python
movie_user_matrix.fillna(0,inplace=True)
```

```python
movie_user_matrix.head()
```

```python
movie_watched = movie_user_matrix["Mohenjo Daro (2016)"]
movie_watched
```
